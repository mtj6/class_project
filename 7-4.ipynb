```python
import pandas as df

def convert_people_cell(cell):
    if cell == 'n.a.':
        return 'sam walton'
    else:
        return cell

df = pd.read_excel('7-4_stock_data.xlsx','Sheet1', converters = {'people' : convert_people_cell})
df
```

```python
df.to_excel('new.xlsx', sheet_name='stocks')
```

```python
df.to_excel('new.xlsx', sheet_name='stocks', startrow=1, startcol=2)
```
